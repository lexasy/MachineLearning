9. Как отсортировать массив по n-му столбцу?

````python
def matrixsort(matrix, n):
    return matrix[matrix[:, n - 1].argsort()]
matrix3 = np.random.randint(0, 5, (3, 3))
print(matrix3)
print(matrixsort(matrix3, 1))
````

10. Как определить, есть ли в данном двумерном массиве пустые столбцы?

````python
def is_axis_null(matrix):
    for i in range(matrix.size // len(matrix)):
        if sum(matrix[:, i]) == 0:
            return True
    return False
matrix4 = np.array([[0, 1, 0], [0, 3, 0], [1, 1, 0]])
print(matrix4)
print(is_axis_null(matrix4))
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
def to_int(input_string):
    string_parts = input_string.split('/') # данные даны в виде "x/y"
    numbers = [int(num) for num in string_parts]
    return sum(numbers) # Сумма систолического и диастолического давлений
print(df[df["Stress Levels"] > np.percentile(df["Stress Levels"], 75)]["Blood Pressure (s/d)"].apply(to_int).mean() > df[df["Stress Levels"] < np.percentile(df["Stress Levels"], 25)]["Blood Pressure (s/d)"].apply(to_int).mean())
````

5. Постройте круговую диаграмму отображающую как ментальное здоровье связано с паттернами сна

````python
pivot_table = df.pivot_table(index="Sleep Patterns", columns="Mental Health Status", aggfunc="size", fill_value=0)
for mental_health_level in pivot_table.columns:
    sizes = pivot_table[mental_health_level]
    labels = pivot_table.index
    colors = ["lightcoral", "gold", "lightskyblue"]
    plt.figure()
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
    plt.title(f'Связь паттерна сна и уровня ментального здоровья {mental_health_level}')
    plt.show()
````

````python
# для всех звуковых частот
vision = df["Vision Sharpness"].values
hearing = df["Hearing Ability (dB)"].values
df1 = pd.DataFrame({"Vision Sharpness": vision, "Hearing Ability": hearing})
sns.lmplot(x="Vision Sharpness", y="Hearing Ability", data=df1, line_kws={"color": "red"})
plt.show()
````

````python
# для звуковых частот выше перцентиля 70
vision = df[df["Hearing Ability (dB)"] > np.percentile(df["Hearing Ability (dB)"], 75)]["Vision Sharpness"].values
hearing = df[df["Hearing Ability (dB)"] > np.percentile(df["Hearing Ability (dB)"], 75)]["Hearing Ability (dB)"].values
df1 = pd.DataFrame({'Vision Sharpness': vision, 'Hearing Ability': hearing})
sns.lmplot(x='Vision Sharpness', y='Hearing Ability', data=df1, line_kws={'color': 'red'})
plt.show()
````

Да, связь есть. В обоих случаях видим, что в целом перед нами представлена линейно убывающая зависимость - с увеличением зрения, уменьшаются слышимые звуковые частоты.
